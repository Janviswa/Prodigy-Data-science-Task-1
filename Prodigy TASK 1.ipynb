```python
ages = np.random.normal(loc=30, scale=10, size=1000)  


genders = ['Male', 'Female', 'Non-Binary']
gender_counts = [450, 520, 30]


fig, axes = plt.subplots(2, 1, figsize=(10, 12))


sns.barplot(x=genders, y=gender_counts, palette='pastel', ax=axes[0])
axes[0].set_title('Distribution of Genders')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Count')

sns.histplot(ages, kde=True, color='teal', ax=axes[1])
axes[1].set_title('Distribution of Ages')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Frequency')


plt.tight_layout()


plt.show()
```
